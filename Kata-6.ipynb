# Ejersicio 2

```python
planets = ["Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", "Uranus", "Neptune"]
planeta = input("Ingresa un nombre de un planeta: ")

i = planets.index(planeta)
print("Los planetas mas cercanos de " + planeta)
print(planets[0:i])
print("Los planetas mas lejos de " + planeta)
print(planets[i + 1:])
```
