# Ejersicio 1

```python
def funcion(tanque1, tanque2, tanque3):
    promedio = (tanque1 + tanque2 + tanque3) / 3
    return "Promedio: {}% tanque1: {} tanque2: {}% tanque3: {}%".format(promedio, tanque1, tanque2, tanque3)

print(funcion(80,70,85))
```

```python
def promedioFun(values):  # Funcion promedio
    total = sum(values)
    numeroItems = len(values)
    return total / numeroItems

def funcion(tanque1, tanque2, tanque3):  # Actualizamos la funcion
    promedio = promedioFun([tanque1, tanque2, tanque3])
    return "Promedio: {}% tanque1: {} tanque2: {}% tanque3: {}%".format(promedio, tanque1, tanque2, tanque3)

print(funcion(88, 76, 70))
```

```python
def reporteMision(pre_launch_time, flight_time, destination, external_tank, main_tank):
    return "Destino: {} Tiempo total de viaje: {} Combustiblee: {}".format(destination, flight_time+pre_launch_time, main_tank+ external_tank)

print(reporteMision(14, 51, "Moon", 200000, 300000))
```

```python
def reporteMision(destination, *minutes, **fuel_reservoirs):  
    return "Mision a {} Tiempo total de viaje: {} Combustiblee: {}".format(destination, sum(minutes), sum(fuel_reservoirs.values())) 

print(reporteMision("Moon", 10, 15, 51, main=300000, external=200000))
```

```python
def mission_report(destination, *minutes, **fuel_reservoirs):  # Funcion copiada ya que no pude resolverla
    main_report = f"""
    Mission to {destination}
    Total travel time: {sum(minutes)} minutes
    Total fuel left: {sum(fuel_reservoirs.values())}
    """
    for tank_name, gallons in fuel_reservoirs.items():
        main_report += f"{tank_name} tank --> {gallons} gallons left\n"
    return main_report

print(mission_report("Moon", 8, 11, 55, main=300000, external=200000))
```
